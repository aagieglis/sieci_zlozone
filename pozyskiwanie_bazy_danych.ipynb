```python
from F1Archive.data.qualy_extractor import QualyExtractor
from F1Archive.data.data_extraction import DataExtractor
import pandas as pd
```

Funkcja zapisująca dane o kierowach w przeciągu wybranych sezonów do pliku csv

```python
def csv_data(last_season, first_season, filename):
    years = [*range(first_season, last_season+1, 1)]
    years = list(reversed(years))
    for i in range(len(years)):
        DX = DataExtractor()
        DX.get_race_urls(years[i])
        DX.seasons_results(print_dataframes=True)
        test_zbior = DX.year_results[str(years[i])]
        df1 = test_zbior[[('Details', 'Driver'), ('Details', 'Car')]]
        df1.insert(2, 'Year', years[i])
        print(df1.tail())
        df1.to_csv(filename, sep=',', encoding='utf-8',mode = 'a', header=False, index=False)
```

```python
csv_data(2009, 2006, "przyklad.csv")
```
